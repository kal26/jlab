```python
# set up a linear fit
def linear(x, *p):
    a, b = p
    return a*x + b

def sinusoid(x, *p):
    a, b, c, d = p
    return a * np.cos(b*x-c) + d

def sinusoid_deriv(x, *p):
    a, b, c, d = p
    return -a * b * np.sin(b*x-c)

def expoential(x, *p):
    a, b, c = p
    return 

def chi_squared(x, fit, sig):
     return np.sum(np.square(x-fit)/np.square(sig))

def reduced_chi_squared(x, fit, sig, N, n):
    return chi_squared(x, fit, sig)/(N-n)

# find the +- chi squared values for fit params
def fit_param_uncerts(fit_function, fit_params, x_vals, y_vals, uncerts):
    orig_chi = chi_squared(y_vals, fit_function(x_vals, *fit_params), uncerts)
    out_chi=(orig_chi, len(x_vals)-len(fit_params), orig_chi/(len(x_vals)-len(fit_params)))
    print('chi '+ str(out_chi[0]))
    print('dof ' + str(out_chi[1]))
    print('reduced ' + str(out_chi[2]))
    
    out_params=list()
    for i in range(len(fit_params)):
        param=fit_params[i]
        change_change=abs(param*10)
        test_change=abs(param/10)
        avg_change=10 #placeholder
        last_avg_change=1000
        while abs(avg_change -1) > .01:
            fit_params_up=fit_params.copy()
            fit_params_down=fit_params.copy()
            fit_params_up[i]=fit_params[i]+test_change
            fit_params_down[i]=fit_params[i]-test_change

            #find chi squareds
            up_chi = chi_squared(y_vals, fit_function(x_vals, *fit_params_up), uncerts)
            down_chi = chi_squared(y_vals, fit_function(x_vals, *fit_params_down), uncerts)
            avg_change = (up_chi+down_chi-2*orig_chi)/2
            
           # print('Got avg change {} with test change {} and change change {} on param {}'.format(avg_change, test_change, change_change, param))
            
            #update for next run
            if avg_change > 1:
                test_change = min(abs(test_change-change_change), test_change)
            else:
                test_change = abs(test_change+change_change)
            if last_avg_change-1<0 != avg_change-1==0:
                change_change = min(change_change/(2), test_change/10)
            else:
                change_change = min(change_change*2, test_change/10)
                
        out_params.append(test_change)
    return out_params, out_chi
```

```python
# custom reading routine
# we into the data ection, but have to skip the first line
def read_file(directory, filename):
    reading=False
    scip=False
    counts=list()
    with open(os.path.join(directory, filename), 'r') as my_file:
        for line in my_file.readlines():
            if '$ROI' in line:
                reading=False
            if reading:
                counts.append(int(line))
            if scip:
                reading=True
                print('reading file ' + filename)
                scip=False
            if '$DATA' in line:
                scip=True
    return np.asarray(counts)           
```

```python
# I want to end up with a dataframe, with columns angle, id, scattercounts, targercounts
num_samples = len(os.listdir(my_folder))//2
angles=[0]*num_samples
ids=[0]*num_samples
scatters=[0]*num_samples
targets=[0]*num_samples
for filename in os.listdir(my_folder):
    # get angle, type, and name    
    degree = filename.split('_')[2]
    scint = filename.split('_')[3].split('.Spe')[0]
    idnum = filename.split('_'+ scint)[0] + filename.split('_'+ scint)[-1]
    idnum=idnum.split('.Spe')[0]
    #print(idnum)
    
    # get the counts from the file
    counts = read_file(my_folder, filename)
    if idnum in ids:
        # we've seen this sample before
        #print('repeat id!')
        spot=ids.index(idnum)
    else:
        # this is a new sample
        spot=ids.index(0)
        ids[spot] = idnum
        angles[spot]= int(degree)
    if scint == 's':
        print('filling scatter data \n')
        scatters[spot]=counts
    else:
        print('filling target data \n')
        targets[spot]=counts
    
    #plt.plot(counts)
    #plt.show()
```

```python
compton = pd.DataFrame(data={'name':ids, 'old_angle':angles, 's':scatters, 't':targets})
compton['angle']=compton['old_angle']
compton.head()
```

```python
# the actual realationship
def scatter_pred(theta, E0=661.6, me=0.5109989461*10**3, c=1):
    return E0/(1+(E0/me)*(1-np.cos(np.deg2rad(theta))))

def target_pred(theta, E0=661.6, me=0.5109989461*10**3, c=1):
    alpha=E0/me
    return E0*(alpha*(1-np.cos(np.deg2rad(theta))))/(1+alpha*(1-np.cos(np.deg2rad(theta))))
```

```python
# translate the channel counts to a list of the enrgies of the signals
def get_energies(counts, scint):
    # get the right energy conversion
    if scint == 's':
        popt=sopt
    else:
        popt=topt
    # do the energy conversion
    channels=list()
    for i, count in enumerate(counts):
        while count > 0:
            channels.append(i)
            count = count-1
    return linear(np.asarray(channels), *popt)

def get_photopeak(row, scint, bins=60, plot=False, minwidth=2, prominance=10):
    # get the predicted values
    if scint == 's':
        pred = scatter_pred(row['old_angle'])
        opred = scatter_pred(180 - row['old_angle'])
    else:
        pred = target_pred(row['old_angle'])
        opred = target_pred(180 - row['old_angle'])
    
    # get the energies
    energy = get_energies(row[scint], scint)
    
    # start scatter plot
    a = plt.hist(energy, bins=bins)
    peaks, props = signal.find_peaks(a[0], prominence=prominance, width=(minwidth,30), height=10)
    if plot:
        plt.plot(a[1][peaks], a[0][peaks],"x")
        plt.axvline(x=pred, color='r', label='orig angle')
        plt.axvline(x=opred, color='b', label='new angle')
    if len(peaks) == 0:
        plt.show()
        print('try lower res')
        peak, ang, d, w  = get_photopeak(row, scint, bins=bins//2, plot=plot, minwidth=0, prominance=prominance//2)
    else:
        peak_id, peak_h = min(enumerate(peaks), key=lambda x: abs(a[1][x[1]]-pred))
        peak_oid, opeak_h = min(enumerate(peaks), key=lambda x: abs(a[1][x[1]]-opred))
        dist=abs(a[1][peak_h]-pred)
        odist=abs(a[1][opeak_h]-opred)
        # see if we should change the angle
        if dist <= odist:
            print('Keep this ones angle!')
            peak = a[1][peaks[peak_id]]
            ang = row['old_angle']
        else:
            print('Change this ones angle!')
            peak = a[1][peaks[peak_id]]
            ang = 180 - row['old_angle']
        #plt.scatter(a[1][peaks[np.argmax(props['prominences'])]], a[0][peaks[np.argmax(props['prominences'])]], s=150, c='r', alpha=.6, label='prominences')
        #plt.scatter(a[1][peaks[np.argmax(props['peak_heights'])]], a[0][peaks[np.argmax(props['peak_heights'])]], s=150, c='m', alpha=.6, label='peak_heights')
        plt.legend()
        #peak = a[1][peaks][-1]
        plt.show()
        d=max(dist, odist)
        w=props['widths'][peak_id]

    return peak, ang, d, w

def get_photopeaks(row, bins=60, plot=False, minwidth=2, prominance=10):
    speak, sang, sd, sw = get_photopeak(row, 's', plot=plot, minwidth=minwidth, bins=bins, prominance=prominance)
    tpeak, tang, td, tw = get_photopeak(row, 't', plot=plot, minwidth=minwidth, bins=bins, prominance=prominance)
    return speak, tpeak, sang, tang, sd, td, sw, tw
    
```

```python
# make the final plot
# statistical uncerts are variences
# systematic uncerts are peak fitting

for key in ['angle']:
    # group the values
    final_angs = list()
    final_svals=list()
    final_suncerts=list()
    sys_suncerts=list()
    final_tvals=list()
    final_tuncerts=list()
    sys_tuncerts=list()
    for ang in set(compton[key]):
        final_angs.append(ang)
        final_svals.append(np.average(compton[compton[key] == ang]['s_energy']))
        final_tvals.append(np.average(compton[compton[key] == ang]['t_energy']))
        # adding to uncerts for statistical varience
        final_suncerts.append(np.sqrt(np.var(compton[compton[key] == ang]['s_energy'])/len(compton[compton[key] == ang])) + final_svals[-1]/20+10)
        final_tuncerts.append(np.sqrt(np.var(compton[compton[key] == ang]['t_energy'])/len(compton[compton[key] == ang]))+ final_tvals[-1]/20+10)
    
        # averaging systematic uncers
        sys_suncerts.append(np.average(compton[compton[key] == ang]['s_e_uncert'])*5)
        sys_tuncerts.append(np.average(compton[compton[key] == ang]['t_e_uncert'])*5)
        
    fig, ax = plt.subplots(1, figsize=(12, 6))
    #final_tuncerts[4]=final_tuncerts[4]/4
    print(final_angs)
    print(final_suncerts)
    print(final_tuncerts)
    

    ax.plot([scatter_pred(t) for t in range(0, 180)],label='scatter prediction',c='blue' )
    ax.plot([target_pred(t) for t in range(0, 180)],label='target prediction',c='red' )

    #plot the t values
    ax.errorbar(final_angs, final_tvals, yerr=final_tuncerts, fmt='.', c='red', ecolor='red', capsize=8, elinewidth=1, markeredgewidth=2, ms=10, label='target')
    # plot the systematic uncerts
    ax.errorbar(final_angs, final_tvals, yerr=sys_tuncerts, fmt='None', ecolor='red', capsize=0, elinewidth=8, label='syst', alpha=0.5)

    # plot the s vals
    ax.errorbar(final_angs, final_svals, yerr=final_suncerts, fmt='.', c='blue', ecolor='blue', capsize=8, elinewidth=1, markeredgewidth=2, ms=10, label='scatter')    
    # plot the systematic uncerts
    ax.errorbar(final_angs, final_svals, yerr=sys_suncerts, fmt='None', ecolor='blue', capsize=0, elinewidth=8, label='syst', alpha=0.5)
    
    
    #ax.errorbar(final_angs, final_tvals, yerr=final_tuncerts, fmt='.', c='blue', ecolor='blue', capsize=8, elinewidth=1, markeredgewidth=2, ms=10, label='scatter')
    #ax.errorbar(final_angs, final_svals, yerr=final_suncerts, fmt='.', c='orange', ecolor='orange', capsize=8, elinewidth=1, markeredgewidth=2, ms=10, label='target')

    
    #ax.scatter(compton[key], compton['s_energy'] + compton['t_energy'], label='total')
    #plt.axhline(y=661.6, color='g')

    plt.title(key)
    plt.legend()
    plt.show
    
    csort = pd.DataFrame(data={'angle':final_angs, 's':final_svals, 't':final_tuncerts, 't_syst':sys_tuncerts, 's_syst':sys_suncerts, 't_stat':final_tuncerts, 's_stat':final_suncerts})
    csort.head()
```
